```python
x=data.drop(["Player Efficiency Rating"],axis=1)

y=data["Player Efficiency Rating"]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
teja=LinearRegression()
teja.fit(x_train,y_train)
```

```python
y_pred=teja.predict(x_test)
```

```python
print("training accuracy:",teja.score(x_train,y_train))
print("testing accuracy:",teja.score(x_test,y_test))
```

```python
y_test=list(y_test)
```

```python
plt.figure(figsize=(20,20))
plt.plot(y_test,label="test",color="blue")
plt.plot(y_pred,label="pred",color="green")
plt.title("Player Efficiency Rating")
plt.legend()
plt.show()
```
